```python
# import
from dotenv import load_dotenv

# load env variables
load_dotenv(override=True)
```

```python
# Check OpenAI API Key
import os
from loguru import logger

OAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OAI_API_KEY:
    logger.info(f"OAI API KEY exists: {OAI_API_KEY[:8]}")
else:
    logger.info(f"OAI API KEY does not exist")
```

```python
# remaining imports
from openai import OpenAI

oai_python_client = OpenAI()
messages = [{
    "role": "user", 
    "content": "what is 2+2?"
}] 

response = oai_python_client.chat.completions.create(
    model='gpt-4o',
    messages=messages
)

logger.info(f"Response: {response.choices[0].message.content}")
```

```python
# harder task
question = "Please propose a hard, challenging question to assess someones's IQ. Respond only with the question."
messages = [{
    "role": "user",
    "content": question
}]

response = oai_python_client.chat.completions.create(
    model='gpt-4o',
    messages=messages
)

question = response.choices[0].message.content
logger.info(f"Question: {question}")
```

```python
messages = [{
    "role": "user",
    "content": question
}]

response = oai_python_client.chat.completions.create(
    model='gpt-4o',
    messages=messages
)

answer = response.choices[0].message.content
logger.info(f"Answer: {answer}")
```

```python
from IPython.display import Markdown, display

display(Markdown(question))
display(Markdown(answer))
```

```python
# Final task
# First round - find an Agentic AI opportunity
question_1 = "Please pick a business area that might be worth exploring for an Agentic AI opportunity."
messages_1 = [{
    "role": "user",
    "content": question_1
}]
response_1 = oai_python_client.chat.completions.create(
    model='gpt-4o',
    messages=messages_1
)
answer_1 = response_1.choices[0].message.content

# Second round - find pain points
question_2 = f"Below you have described a business are worth using Agentic AI. Please provide some pain-points in that industry and areas in which AI agents can thrive. Description: {response_1}"
messages_2 = [{
    "role": "user",
    "content": question_2
}]
response_2 = oai_python_client.chat.completions.create(
    model='gpt-4o',
    messages=messages_2
)
answer_2 = response_2.choices[0].message.content

# Third round - propose agentic solution
question_3 = f"Below you'll find two summaries which are to describe business are worth exploring by Agentic AI and it's pain-points. Analyze this data and propose a solution based on AI agents to solve this task as efficiently as possible.\nDescription: {response_1}\nPain-points: {response_2}"
messages_3 = [{
    "role": "user",
    "content": question_3
}]
response_3 = oai_python_client.chat.completions.create(
    model='gpt-4o',
    messages=messages_3
)
answer_3 = response_3.choices[0].message.content
```

```python
from IPython.display import Markdown, display

display(Markdown(question_1))
display(Markdown(answer_1))
display(Markdown(answer_2))
display(Markdown(answer_3))
```
